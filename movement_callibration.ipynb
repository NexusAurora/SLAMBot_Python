```python
import numpy as np
import numpy as np
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt
```

```python
weights = {
}
```

```python
#! Forward [speed - 100]
x = np.array([0, 32, 33.5, 32.2, 34.2, 11.4, 10.3, 20.5, 20]) #cm
y = np.array([0, 1000, 1000, 1000, 1000, 500, 500, 700, 700])

# Define a polynomial model
def model(x, a, b, c, d):
    return a * x**3 + b * x**2 + c * x + d

# Fit the model to the data
popt, _ = curve_fit(model, x, y)

# Print the parameters
print(popt)

#store the weights
weights['forward'] = popt

# Make a prediction for one value
x_new = np.array([100])
print(model(x_new, *popt))

# Plot the data and the fitted model (line)
plt.scatter(x, y)
x_line = np.arange(0, 40, 0.1)
y_line = model(x_line, *popt)
plt.plot(x_line, y_line, color='red')
plt.show()
```
